```python
tokenizer.pad_token = tokenizer.eos_token
```

```python
def inference (model, text, tokenizer, max_input_tokens = 1000, max_output_tokens = 100):
  encodings = tokenizer(
      text,
      return_tensors = 'pt',
      truncation = True,
      padding = True,
      max_length = max_input_tokens
  )


  input_ids = encodings['input_ids']
  attention_mask = encodings['attention_mask']

  generated_tokens = model.generate(
      input_ids = input_ids,
      attention_mask = attention_mask,
      max_new_tokens = max_output_tokens
  )

  generated_text = tokenizer.batch_decode(generated_tokens, skip_special_tokens = True)
  answer = generated_text[0]

  return generated_text
```

```python
input = "Please write a short and precise caption for clear lungs"
answer = inference(model, input, tokenizer)
print("Question: ", input)
print("Answer: ", answer)
```
